```python
import pandas as pd
import os

def combine_ingredient_columns(df):
    # Combine all ingredient columns starting from the third column into one column
    ingredient_columns = df.columns[2:]  # Columns 3 and beyond
    df['ingredients'] = df[ingredient_columns].apply(lambda row: ' , '.join(row.dropna().astype(str)), axis=1)
    # Keep only the 'name', 'url', and 'ingredients' columns
    df = df[['name', 'url', 'ingredients']]
    return df

# Manually define the list of CSV file names with the correct directory path
file_names = [
    '../Web Scraping - CS/recipes_a_full.csv', '../Web Scraping - CS/recipes_b.csv', 
    '../Web Scraping - CS/recipes_c_full.csv', '../Web Scraping - CS/recipes_d_full.csv', 
    '../Web Scraping - CS/recipes_e_full.csv', '../Web Scraping - CS/recipes_f_full.csv', 
    '../Web Scraping - CS/recipes_g_full.csv', '../Web Scraping - CS/recipes_h_full.csv', 
    '../Web Scraping - CS/recipes_i_full.csv', '../Web Scraping - CS/recipes_j_full.csv', 
    '../Web Scraping - CS/recipes_k_full.csv', '../Web Scraping - CS/recipes_l_full.csv', 
    '../Web Scraping - CS/recipes_m_full.csv', '../Web Scraping - CS/recipes_n_full.csv', 
    '../Web Scraping - CS/recipes_o_full.csv', '../Web Scraping - CS/recipes_p_full.csv', 
    '../Web Scraping - CS/recipes_q_full.csv', '../Web Scraping - CS/recipes_r_full.csv', 
    '../Web Scraping - CS/recipes_s_full.csv', '../Web Scraping - CS/recipes_t_full.csv', 
    '../Web Scraping - CS/recipes_u_full.csv', '../Web Scraping - CS/recipes_v_full.csv', 
    '../Web Scraping - CS/recipes_w_full.csv', '../Web Scraping - CS/recipes_xyz_full.csv'
]  # Add more file names as needed

# Initialize an empty dataframe to hold all combined data
combined_df = pd.DataFrame()

# Process each file
for file_name in file_names:
    # Read the CSV file
    df = pd.read_csv(file_name)
    
    # Combine the ingredient columns
    df_combined = combine_ingredient_columns(df)
    
    # Append the combined dataframe to the master dataframe
    combined_df = pd.concat([combined_df, df_combined], ignore_index=True)

# Display the first 50 rows of the combined dataframe
combined_df.head(50)
```

```python
import re
import pandas as pd

def clean_ingredients(ingredients):
    # Remove special characters using regular expressions
    clean_ingredients = re.sub(r'[^\w\s,|]', '', ingredients)
    return clean_ingredients

def capitalize_words(ingredients):
    # Capitalize only the first letter of fully capitalized words
    words = ingredients.split()
    capitalized_words = [word.capitalize() if word.isupper() else word for word in words]
    return ' '.join(capitalized_words)

def remove_deselect_all(ingredients):
    # Remove "Deselect All" from ingredients
    clean_ingredients = ingredients.replace('Deselect All ,', '').replace('Deselect All', '').strip()
    return clean_ingredients

def clean_name(name):
    # Remove special characters from name
    clean_name = re.sub(r'[^\w\s]', '', name)
    return clean_name

def count_ingredients(ingredients):
    # Count the number of ingredients in the recipe
    return len(ingredients.split(','))

# Apply the cleaning functions to the ingredients column
combined_df['ingredients'] = combined_df['ingredients'].apply(clean_ingredients)
combined_df['ingredients'] = combined_df['ingredients'].apply(capitalize_words)
combined_df['ingredients'] = combined_df['ingredients'].apply(remove_deselect_all)
combined_df['name'] = combined_df['name'].apply(clean_name)

# Filter recipes with more than 6 and less than 22 ingredients
combined_df['ingredient_count'] = combined_df['ingredients'].apply(count_ingredients)
filtered_df = combined_df[(combined_df['ingredient_count'] > 10) & (combined_df['ingredient_count'] < 25)]

# Drop the ingredient count column after filtering
filtered_df = filtered_df.drop(columns=['ingredient_count'])

# Set pandas option to display more of the ingredients column
pd.options.display.max_colwidth = 1000  # Adjust the value as needed

# Display the first 50 rows of the cleaned and filtered dataframe
filtered_df.head(50)
```
